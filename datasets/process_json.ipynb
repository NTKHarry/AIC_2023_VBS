$\text{Đoạn này là bắt đầu tách nè, hàm bên dưới là chia 1 video thành nhiều frame}$

```python
import cv2
import numpy as np
def extract_specific_frames(video_path, frame_rate, threshold = 30):
    vidcap = cv2.VideoCapture(video_path)
    frames = []
    frame = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
    cnt = 0
# Lấy độ phân giải của video
    frame_width = int(vidcap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # In độ phân giải của video
    #print(f'Độ phân giải của video: {frame_width}x{frame_height}')
    prev_frame = None
    t = 0
    for idx in range(0,frame,frame_rate):
        vidcap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        success, image = vidcap.read()
    # if success:
    #     frames.append(image)
        if success:
            if prev_frame is None:
                prev_frame = image
                frames.append(image)
                continue
            diff = cv2.absdiff(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY))
            non_zero_count = np.count_nonzero(diff)
            cnt += non_zero_count
            t += 1
            if non_zero_count > frame_width*frame_height*0.95:
                frames.append(image)
                prev_frame = image
    vidcap.release()
    #print(cnt / t)
    return frames

frames = extract_specific_frames("/Users/vothanhnhan/Desktop/gift/AIC/datasets/video_news/60 Giây Chiều - Ngày 01⧸02⧸2023 - HTV Tin Tức Mới Nhất [blWiOg3lsEE].mp4",30)
```
